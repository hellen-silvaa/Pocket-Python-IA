# Etapa 1 – Leitura e Introdução

1. O que é Data Science.

- Ciência de dados é a prática multidisciplinar de coletar, limpar, analisar e interpretar dados com ferramentas de estatística, programação e aprendizado de máquina para transformar dados brutos em insights úteis para a tomada de decisões. Em termos práticos, responde ao que aconteceu, por que aconteceu, o que provavelmente acontecerá e que ações podem ser tomadas. fazendo analogia é como cozinhar você reúne ingredientes (dados), prepara e limpa os ingredientes (pré-processamento), segue uma receita ou inventa uma (modelo/algoritmo) e ajusta o tempero até o prato ficar bom (validação e interpretação).
para mim data science é o processo que transforma grandes volumes de informação em respostas acionáveis que orientam decisões estratégicas.

2. Principais bibliotecas do Python para análise de dados.

- pandas — Manipulação e análise de tabelas (DataFrame/Series).
- NumPy — Arrays e operações numéricas rápidas; base para cálculos científicos.
- Matplotlib — Biblioteca base para gráficos 2D customizáveis.
- Seaborn — Gráficos estatísticos prontos e estéticos (sobre Matplotlib).
- scikit‑learn — Algoritmos clássicos de machine learning e pré‑processamento.
- SciPy — Funções científicas (álgebra linear, otimização, integração).
- Statsmodels — Modelos estatísticos e testes formais (regressão, ANOVA, séries).
- Plotly — Visualizações interativas e publicáveis na web.
- SQLAlchemy — Acesso e integração com bancos de dados SQL a partir do Python.
- BeautifulSoup — Parser de HTML/XML para extrair dados de páginas (web scraping leve).

3. Etapas de um projeto em Data Science.
- 1 Definir o problema — Esclarecer objetivo, perguntas a responder, métricas de sucesso e stakeholders. (Analogia: desenhar a planta da casa antes de construir.)
- 2 Coletar os dados — Reunir dados relevantes de fontes internas/externas, APIs, arquivos e bancos.
- 3 Pré‑processar os dados — Limpar, tratar valores ausentes/outliers, transformar e escalar; preparar para análise.
- 4 Explorar e analisar (EDA) — Visualizar, resumir estatísticas e identificar padrões e correlações.
- 5 Modelar — Selecionar e treinar algoritmos (regressão, classificação, cluster, etc.) e ajustar hiperparâmetros.
- 6 Avaliar e validar — Medir performance com métricas e validação cruzada; checar vieses e robustez.
- 7 Implementar e monitorar — Colocar em produção, integrar com sistemas e acompanhar desempenho ao longo do tempo.

# Etapa 2 – Criação do Dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
#criação do dataset
dados = {
    "Nome": ["Maria Helena", "Hellen Silva", "Angelica Souza", "Maria Madalena", "Gabriela Rocha"],
    "Idade": [28, 22, 22, 24, 19],
    "Salário": [3500.00, 56000.50, 2900.00, 7200.00, 4200.75],
    "Cidade": ["Rio de Janeiro", "São Paulo", "Belo Horizonte", "Curitiba", "Piauí"]
}

df = pd.DataFrame(dados)
```

```python
# exibir 5 primeiras linhas e mostrar cada coluna
print("---------- 5 primeiras linhas ----------")
display(df.head(5))

print("\n---------- Tipos das colunas ----------")
print(df.dtypes)
```

```python
# calcular a média salarial
media_salarios = df["Salário"].mean()
print(f"Média dos salários: R$ {media_salarios:,.2f}")
```

# Etapa 3 – Análise Exploratória

```python
# pessoas com salário > R$ 4.000
print("Pessoas com salário > R$ 4.000:")
display(df[df["Salário"] > 4000.0])

# ordenação por idade
ordenado_por_idade = df.sort_values(by="Idade", ascending=True).reset_index(drop=True)
print("\nDados ordenados por Idade:")
display(ordenado_por_idade)

# coluna "Salário Anual" (12 x salário)
df["Salário Anual"] = df["Salário"] * 12
print("\nDataFrame com nova coluna 'Salário Anual':")
display(df)

# dtypes atualizados
print("\nTipos atualizados das colunas:")
print(df.dtypes)
```

# Etapa 4 – Visualização e Conclusão

```python
plt.figure(figsize=(9,5))
bars = plt.bar(df["Nome"], df["Salário"], color='skyblue', edgecolor='black')
plt.title("Salários por Pessoa")
plt.xlabel("Nome")
plt.ylabel("Salário (R$)")
plt.xticks(rotation=20, ha='right')

# Anotar valores acima das barras
for bar in bars:
    h = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, h + 120, f"R$ {h:,.2f}", ha='center', va='bottom', fontsize=9)

plt.tight_layout()
plt.show()
```

```python
#Conclusão:

#Ao analisar os dados notei variação salarial significativa e a presença de um outlier (que sou eu) Hellen Silva com R$ 56.000,50 que deve ser verificado.
#A maioria das pessoas é jovem (idades entre 19 e 28), indicando um perfil de início de carreirae são mulheres
#A média pode ser distorcida pelo outlier; calcular mediana e desvio padrão ajuda a entender melhor a distribuição dos dados.
#Para a empresa, identificar quem está acima ou abaixo da mediana é útil para políticas de retenção, promoção e alinhamento salarial.
#A coluna "Salário Anual" permite estimar custos trabalhistas anuais e planejar o orçamento.
```
