** PLN Processamento de Linguagem Natural e Modelos de Linguagem para Mineração de textos**



```python
import pandas as pd
from collections import Counter
import matplotlib.pyplot as plt
```

Exemplo 1: vamos fazer uma análise simples para descobrir as palavras mais frequentes em um conjunto de comentários e clientes.

```python
comentarios = [
"O produto é excelente, adorei a qualidade!"
"A entrega demorou muito, estou extremamente insatisfeito."
"Gostei muito do atendimento, mas o produto veio com defeito."
"Qualidade impecável, o produto não funciona como esperado."
                                                                                                                                                                                                                                        "
]
```

```python
#Juntar todos os comentários em um único texto
texto_completo = " ".join(comentarios)
```

```python
#Sepração simples, dividir o texto em palavras (tokenização)

palavras = texto_completo.lower().split()
```

```python
#Contagem de frequência das palavras

contagem_palavras = Counter(palavras)
```

```python
#Palavras em comum

mais_comuns = contagem_palavras.most_common(10)
```

```python
#Visualização

df_mais_comun = pd.DataFrame(mais_comuns, columns=['Palavra', 'Frequencia'])

plt.figure(figsize=(10,6))
plt.bar(df_mais_comun['Palavra'], df_mais_comun['Frequencia'], color='skyblue')
plt.xlabel('Palavras')
plt.ylabel('Frequência')
plt.title('Palavras Mais Frequentes')
plt.xticks(rotation=45)
plt.show()
```
